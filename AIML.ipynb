# **Matrix opration**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# dot product
v1 = np.array([2, 3])
v2 = np.array([4, 4])

res = np.dot(v1, v2)             # dot product of v1 and v2 vector
print(res)
```

```python
v3 = np.array([40, 20, 35])
v4 = np.array([53, 24, 68])

res = np.dot(v3, v4)           # dot product of v3 and v4 vector
print(res)
```

```python
# cross product
v1 = np.array([2, 3])
v2 = np.array([4, 4])
res = np.cross(v1, v2)         # cross product of v1 and v2 vector
print(res)

v3 = np.array([5, 10, 20])
v4 = np.array([18, 32, 50])
res = np.cross(v3, v4)         # cross product of v3 and v4 vector
print(res)
```

# **KNN (k-Nearest Neighbor) algorithm**

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# load iris datasets
iris = load_iris()

x = iris.data
y = iris.target

# spliting x_tarin, x_test, y_train, y_test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)


# scaling independent variable
from sklearn.preprocessing import StandardScaler

sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.fit_transform(x_test)
```

# **Hierarchical clustering**
* Aggromerative: It is an bottom-up approach
* Divisive: It is an top-down approach

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler

x, _ = make_blobs(n_samples=100, centers=3, random_state=42)

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
z = linkage(x_scaled, method='ward')  # ward calculate minimum varians  (shortest distance)
```

```python
# printing dandrogram
plt.figure(figsize = (10, 7))
dendrogram(z)
plt.title('Dendrogram for Hierarchical clustering')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python
# aggrometic clustering
agg_clustering = AgglomerativeClustering(n_clusters = 5, linkage='ward')
labels = agg_clustering.fit_predict(x_scaled)      # traing and predict
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x_scaled[:, 0], x_scaled[:, 1], c=labels, cmap='rainbow')
plt.title('Agglomerative Clustering result')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

# **Decision tree**
> Decision tree is a machine learning algorithm based on supervised learning that can be used for both regression and classification problems.

> The goal is to built up a classificer mode that can predict the class or values of the target variable.

---

* Root: The heighes node in the tree represents the original choice or feature from which the tree branches.

* Criteria: Criteria is used for find target variable in best case it is also called impority criteria.

* Entropy: The common way for find out the impurity criteria is called Entropy. the entropy value can be very from 0 - 1.

  - calculation: P = sumation of (Px.log(Px))
  - type of entropy:
    1. Low entropy: high information gain ( Pure subset )
    2. Heigh entropy: low information gain ( Mixed subset )

* Gini impurity: internal wowrking for find the subset in decision tree is called Gini impurity. Gini impurity is same of the Entropy in Decision tree.
  - Value of Gini Index between 0 - 0.5
  - Gini index: Gini(P) = 1 - sumation of (Px^2)
